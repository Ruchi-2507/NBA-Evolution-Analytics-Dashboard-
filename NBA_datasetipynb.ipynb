```python
# =========================
# 1. GAMES TABLE
# =========================
games = games.drop(columns=[
    'GAME_STATUS_TEXT'
], errors='ignore')


# =========================
# 2. GAME_DETAILS TABLE
# =========================
game_details = game_details.drop(columns=[
    'NICKNAME',
    'COMMENT',
    'TEAM_ABBREVIATION',
    'TEAM_CITY',
    'START_POSITION',
    'MIN','FGM','FGA','FG_PCT',
    'FG3M','FG3A','FG3_PCT',
    'FTM','FTA','FT_PCT',
    'OREB','DREB','PF'
], errors='ignore')

# Keep only required columns (FINAL OPTIMIZATION)
game_details = game_details[
    ['GAME_ID','PLAYER_ID','TEAM_ID',
     'PTS','AST','REB','STL','BLK','TO','PLUS_MINUS']
]


# =========================
# 3. RANKINGS TABLE
# =========================
ranking = ranking.drop(columns=[
    'RETURNTOPLAY',
    'LEAGUE_ID', 'HOME_RECORD', 'ROAD_RECORD' ], errors='ignore')


# =========================
# 4. TEAMS TABLE
# =========================
teams = teams.drop(columns=[
    'LEAGUE_ID','MIN_YEAR','MAX_YEAR',
    'ARENA','ARENACAPACITY',
    'OWNER','GENERALMANAGER',
    'HEADCOACH','DLEAGUEAFFILIATION'
], errors='ignore')

# Keep only useful columns
teams = teams[
    ['TEAM_ID','CITY','NICKNAME','YEARFOUNDED']
]


# =========================
# 5. PLAYERS TABLE
# =========================
players = players[['PLAYER_ID','PLAYER_NAME','SEASON']]


# =========================
# FINAL CHECK
# =========================
print("Games:", games.shape)
print("Game Details:", game_details.shape)
print("Ranking:", ranking.shape)
print("Teams:", teams.shape)
print("Players:", players.shape)
```

```python
games['TOTAL_POINTS'] = games['PTS_home'] + games['PTS_away']
```

```python
games['POINT_DIFF'] = abs(games['PTS_home'] - games['PTS_away'])
```

```python
games['WINNER_TEAM_ID'] = np.where(
    games['HOME_TEAM_WINS'] == 1,
    games['TEAM_ID_home'],
    games['TEAM_ID_away']
)
```

```python
game_details['EFFICIENCY'] = (
    game_details['PTS'] +
    game_details['REB'] +
    game_details['AST'] +
    game_details['STL'] +
    game_details['BLK'] -
    game_details['TO']
)
```

```python
game_details['IMPACT'] = game_details['PTS'] + game_details['PLUS_MINUS']
```

```python
game_details['OFFENSE'] = game_details['PTS'] + game_details['AST']
```

```python
ranking['WIN_RATIO'] = ranking['W'] / (ranking['W'] + ranking['L'])
```

```python
games['CLOSE_GAME'] = games['POINT_DIFF'] <= 5
```

```python
games['HIGH_SCORING'] = games['TOTAL_POINTS'] > 220
```

```python
ranking['STANDINGSDATE'] = pd.to_datetime(
    ranking['STANDINGSDATE'],
    dayfirst=True,
    errors='coerce'
)
```

```python
Home_stats = games.groupby(['SEASON','TEAM_ID_home']).agg(
    TOTAL_HOME_GAMES=('GAME_ID','count'),
    AVG_HOME_POINTS=('PTS_home','mean'),
    TOTAL_HOME_WINS=('HOME_TEAM_WINS','sum')
).reset_index()
```

```python
Away_stats = games.groupby(['SEASON','TEAM_ID_away']).agg(
    TOTAL_AWAY_GAMES=('GAME_ID','count'),
    AVG_AWAY_POINTS=('PTS_away','mean'),
    TOTAL_AWAY_WINS=('HOME_TEAM_WINS', lambda x: (1 - x).sum())
).reset_index()
```

```python
team_summary = pd.merge(
    Home_stats,
    Away_stats,
    left_on=['SEASON','TEAM_ID_home'],
    right_on=['SEASON','TEAM_ID_away'],
    how='outer'
)
```

```python
team_summary['TEAM_ID'] = team_summary['TEAM_ID_home'].combine_first(team_summary['TEAM_ID_away'])

team_summary['TOTAL_GAMES'] = team_summary['TOTAL_HOME_GAMES'].fillna(0) + team_summary['TOTAL_AWAY_GAMES'].fillna(0)

team_summary['TOTAL_WINS'] = team_summary['TOTAL_HOME_WINS'].fillna(0) + team_summary['TOTAL_AWAY_WINS'].fillna(0)

team_summary['WIN_PCT'] = team_summary['TOTAL_WINS'] / team_summary['TOTAL_GAMES']

team_summary['AVG_POINTS'] = (
    team_summary['AVG_HOME_POINTS'].fillna(0) +
    team_summary['AVG_AWAY_POINTS'].fillna(0)
) / 2
```

```python
# =========================
# STEP 1: BASE AGGREGATION
# =========================
player_summary = game_details.groupby(['PLAYER_ID']).agg(
    AVG_POINTS=('PTS','mean'),
    AVG_AST=('AST','mean'),
    AVG_REB=('REB','mean'),
    AVG_EFFICIENCY=('EFFICIENCY','mean'),
    TOTAL_GAMES=('GAME_ID','count')
).reset_index()


# =========================
# STEP 2: CONSISTENCY (STD)
# =========================
player_std = game_details.groupby('PLAYER_ID')['PTS'].std().reset_index()
player_std.rename(columns={'PTS':'POINTS_STD'}, inplace=True)

player_summary = pd.merge(
    player_summary,
    player_std,
    on='PLAYER_ID',
    how='left'
)


# =========================
# STEP 3: CLEAN PLAYERS TABLE
# =========================
players_unique = players[['PLAYER_ID','PLAYER_NAME']].drop_duplicates()


# =========================
# STEP 4: REMOVE OLD NAME COLS (AVOID ERRORS)
# =========================
player_summary = player_summary.drop(columns=[
    'PLAYER_NAME','PLAYER_NAME_x','PLAYER_NAME_y'
], errors='ignore')


# =========================
# STEP 5: MERGE PLAYER NAMES
# =========================
player_summary = pd.merge(
    player_summary,
    players_unique,
    on='PLAYER_ID',
    how='left'
)


# =========================
# STEP 6: FILTER IMPORTANT PLAYERS
# =========================
player_summary = player_summary[player_summary['TOTAL_GAMES'] > 20]


# =========================
# STEP 7: CREATE PLAYER CATEGORY
# =========================
player_summary['PLAYER_LEVEL'] = pd.cut(
    player_summary['AVG_POINTS'],
    bins=[0,10,20,30,100],
    labels=['Low','Average','Good','Elite']
)


# =========================
# STEP 8: SORT FOR DASHBOARD
# =========================
player_summary = player_summary.sort_values(
    by='AVG_POINTS',
    ascending=False
)


# =========================
# STEP 9: FINAL CHECK
# =========================
print(player_summary.head())

print("Unique Players:", player_summary['PLAYER_ID'].nunique())
print("Total Rows:", len(player_summary))


# =========================
# STEP 10: SAVE FILE
# =========================
player_summary.to_csv("player_summary.csv", index=False)
```
